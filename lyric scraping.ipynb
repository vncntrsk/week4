```python
lyrics_clean = []
#for j in range(len(lyrics_final)):
#    lyrics_txt = lyrics_final[j]
#    lyrics_clean.append(re.sub('\[Chorus\]|\n|\\\\|\"', ' ', lyrics_txt))
for j in range(len(lyrics_final)):
    lyrics_txt = lyrics_final[j]
    lyrics_clean.append(re.findall(word, lyrics_txt.lower()))
```

```python
#https://coderwall.com/p/rcmaea/flatten-a-list-of-lists-in-one-line-in-python
#The list of lists
list_of_lists = [range(len(lyrics_clean)), range(10)]
#flatten the lists
flattened_list = [y for x in lyrics_clean for y in x]
```

```python
cv = CountVectorizer()
vec = cv.fit_transform(flattened_list)
```

```python
cv.vocabulary_.get(u'algorithm')
```

```python
cv.vocabulary_.keys()
```

```python
words = sorted(cv.vocabulary_.keys())
len(words)
```

```python
lda = LatentDirichletAllocation(n_components=10, max_doc_update_iter=500)
lda.fit(vec)
```

```python
c = lda.components_
c.shape # topics x words
```

```python
ctrans = c.T
ctrans.shape
```

```python
df3 = pd.DataFrame(ctrans, index=words)
df3.tail()
```
